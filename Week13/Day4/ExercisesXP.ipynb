# 🌟 Exercise 2 : Scraping robots.txt from Wikipedia

```python
from urllib.request import Request, urlopen

# URL of robots.txt
url = "https://en.wikipedia.org/robots.txt"

# Create request with headers
request = Request(
    url,
    headers={"User-Agent": "Mozilla/5.0"}
)

# Open URL and read content
response = urlopen(request)

# Decode content
robots_txt = response.read().decode("utf-8")

# Print robots.txt
print(robots_txt)
```

**🌟 Exercise 3 : Extracting Headers from Wikipedia’s Main Page**


```python
# Import libraries
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

# Wikipedia Main Page URL
url = "https://en.wikipedia.org/wiki/Main_Page"

# Create request with User-Agent
request = Request(
    url,
    headers={"User-Agent": "Mozilla/5.0"}
)

# Open URL and read HTML
response = urlopen(request)
html = response.read()

# Parse HTML with BeautifulSoup
soup = BeautifulSoup(html, "html.parser")

# Find all header tags (h1 to h6)
headers = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])

# Display headers
print("Wikipedia Header Tags:\n")

for header in headers:
    print(header.name + ":", header.text.strip())
```

**🌟 Exercise 5 : Analyzing US-CERT Security Alerts**

```python
# Import libraries
import requests
from bs4 import BeautifulSoup
from datetime import datetime

# Current year
current_year = datetime.now().year

# URL
url = "https://www.cisa.gov/news-events/cybersecurity-advisories?f%5B0%5D=advisory_type%3A93"

# Send request
response = requests.get(
    url,
    headers={"User-Agent": "Mozilla/5.0"}
)

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Find all advisory dates on the page
dates = soup.find_all("time")

# Counter
count = 0

# Count alerts from current year
for date in dates:
    if str(current_year) in date.text:
        count += 1

# Display result
print(f"Number of US-CERT security alerts in {current_year}: {count}")
```

**🌟 Exercise 6 : Scraping Movie**

```python
# Install libraries if needed
!pip install beautifulsoup4 requests

# Import libraries
import requests
from bs4 import BeautifulSoup
import random

# IMDb URL
url = "https://www.imdb.com/es-es/list/ls091294718/"

# Headers to avoid blocking
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Send request
response = requests.get(url, headers=headers)

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Find all movie items
movies = soup.find_all("li", class_="ipc-metadata-list-summary-item")

print("Movies found:", len(movies))

# Select up to 10 random movies
random_movies = random.sample(movies, min(10, len(movies)))

# Loop through movies
for i, movie in enumerate(random_movies, start=1):

    # Movie title
    title_tag = movie.find("h3")
    title = title_tag.text if title_tag else "No title"

    # Movie year
    year_tag = movie.find("span", class_="sc-14dd939d-6")
    year = year_tag.text if year_tag else "No year"

    # Movie summary
    summary_tag = movie.find("div", class_="ipc-html-content-inner-div")
    summary = summary_tag.text.strip() if summary_tag else "No summary"

    # Print results
    print(f"\nMovie {i}")
    print("Name:", title)
    print("Year:", year)
    print("Summary:", summary)
    print("-" * 50)
```
