```python
# Data Loading and Cleaning

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load dataset
df = pd.read_csv("Palmer_Penguins_Dataset.csv")

# Basic info
print(df.shape)
print(df.columns)
print(df.head())

# Check missing values
print(df.isnull().sum())

# Drop rows with missing values (simple)
df = df.dropna()

print("After dropna:", df.shape)
```

```python
# Exploratory Data Analysis (EDA):

print(df.describe())
print(df["species"].value_counts())
print(df["island"].value_counts())
print(df["sex"].value_counts())
```

```python
# Visualization Tasks:

# Species Distribution
plt.figure()
sns.countplot(data=df, x="species")
plt.title("Penguin Species Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# Flipper Length vs. Body Mass sactter plot
plt.figure()
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Flipper Length vs Body Mass")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()

# Island Distribution with a pie chart
island_counts = df["island"].value_counts()

plt.figure()
plt.pie(island_counts.values, labels=island_counts.index, autopct="%1.1f%%")
plt.title("Penguins by Island")
plt.show()

# Sex Distribution within Species bar charts
plt.figure()
sns.countplot(data=df, x="species", hue="sex")
plt.title("Sex Distribution within Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# Heatmap of Correlations between numerical variables

numeric_cols = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Columns)")
plt.show()

# Boxplots for Measurements

plt.figure()
sns.boxplot(data=df, x="species", y="flipper_length_mm")
plt.title("Flipper Length by Species")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.show()

plt.figure()
sns.boxplot(data=df, x="species", y="body_mass_g")
plt.title("Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()

plt.figure()
sns.boxplot(data=df, x="species", y="bill_length_mm")
plt.title("Bill Length by Species")
plt.xlabel("Species")
plt.ylabel("Bill Length (mm)")
plt.show()

plt.figure()
sns.boxplot(data=df, x="species", y="bill_depth_mm")
plt.title("Bill Depth by Species")
plt.xlabel("Species")
plt.ylabel("Bill Depth (mm)")
plt.show()

# Histograms of Measurements

plt.figure()
sns.histplot(data=df, x="bill_length_mm", kde=True, bins=20)
plt.title("Distribution of Bill Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Count")
plt.show()

plt.figure()
sns.histplot(data=df, x="bill_depth_mm", kde=True, bins=20)
plt.title("Distribution of Bill Depth")
plt.xlabel("Bill Depth (mm)")
plt.ylabel("Count")
plt.show()

plt.figure()
sns.histplot(data=df, x="flipper_length_mm", kde=True, bins=20)
plt.title("Distribution of Flipper Length")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Count")
plt.show()

plt.figure()
sns.histplot(data=df, x="body_mass_g", kde=True, bins=20)
plt.title("Distribution of Body Mass")
plt.xlabel("Body Mass (g)")
plt.ylabel("Count")
plt.show()
```

```python
# Insightful Visualizations:

# Subplots comparing different aspects of the data.
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.countplot(data=df, x="species", ax=axes[0, 0])
axes[0, 0].set_title("Species Count")

sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species", ax=axes[0, 1], legend=False)
axes[0, 1].set_title("Flipper vs Mass")

sns.histplot(data=df, x="body_mass_g", bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Body Mass Distribution")

sns.boxplot(data=df, x="species", y="body_mass_g", ax=axes[1, 1])
axes[1, 1].set_title("Body Mass by Species")

plt.tight_layout()
plt.show()
```

**Summary:**

The analysis of the Palmer Penguins dataset revealed clear differences between species.

Gentoo penguins have the highest body mass and longest flippers. Adelie penguins are smaller, and Chinstrap penguins fall in between.

There is a strong positive relationship between flipper length and body mass, meaning larger penguins tend to have longer flippers.

The heatmap showed strong correlations between several body measurements, confirming consistent physical relationships.

Overall, data visualization helped identify patterns, differences between species, and relationships between physical measurements.
