**Daily Challenge: NumPy, Pandas, and Matplotlib Integration**

**Challenge Description:**

**Scenario:**
You are a data analyst working with a dataset of global weather. Your task is to analyze temperature trends and visualize the results.

**Tasks:**

**1. Data Preparation:**
* Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.
* Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.
* Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.
* Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.

**2. Data Analysis:**
* Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).
* Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.
* Calculate the annual average temperature for each city.
* Identify the city with the highest and lowest average temperature for the year.

**3. Data Visualization**

```python
# 1.Data Preparation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reproducibility
np.random.seed(42)

# Generate the temperature data
temperature_data = np.random.uniform(-5, 35, (10, 12))

# Cities and months of the year
cities = [
    "Tel Aviv", "Buenos Aires", "New York", "London", "Tokyo",
    "Paris", "Berlin", "Sydney", "Toronto", "Cape Town"
]

months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

df = pd.DataFrame(temperature_data, index=cities, columns=months)

df.head()
```

```python
# 2. Data Analysis

# annual average temperature
annual_avg = df.mean(axis=1)

annual_avg

# cities with the highest and lowest average temperature
highest_city = annual_avg.idxmax()
lowest_city = annual_avg.idxmin()

print("Highest average temperature:", highest_city)
print("Lowest average temperature:", lowest_city)
```

```python
# 3. Data Visualization

plt.figure(figsize=(12,6))

for city in df.index:
    plt.plot(months, df.loc[city], alpha=0.6)

plt.title("Monthly Temperature Trends by City")
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.legend(df.index, bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

**Summary:**

The city with the highest average annual temperature is Buenos Aires.

The city with the lowest average annual temperature is Sydney.

The visualization shows seasonal variations across all cities.

Some cities exhibit higher variability in monthly temperatures compared to others.
