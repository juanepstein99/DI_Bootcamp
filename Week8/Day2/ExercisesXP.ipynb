**üåü Exercise 2 : Creating a Line Plot for Temperature Variation**

Instructions:
* Create a simple line plot using Matplotlib that represents the temperature variations over a week.
* Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.


```python
import matplotlib.pyplot as plt

# Data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Line plot
plt.plot(days, temperatures)

# Labels and title
plt.xlabel('Day')
plt.ylabel('Temperature (¬∞F)')
plt.title('Temperature Variation Over a Week')

plt.show()
```

**üåü Exercise 4 : Data Visualisation**

Instructions:

Using the dataset: sales_data.csv

* Calculate the total quantity of products sold by the company.
* Identify the category that had the highest revenue and calculate how much revenue it generated.
* Determine the average revenue per sale in the dataset.
* Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
import pandas as pd

df = pd.read_csv("sales_data.csv")

# Convert the date column to datetime
df["date"] = pd.to_datetime(df["date"])

# Total quantity of products sold
total_quantity = df["quantity"].sum()
print("Total quantity sold:", total_quantity)

# Category with the highest total revenue
revenue_by_category = df.groupby("category")["revenue"].sum()
top_category = revenue_by_category.idxmax()
top_category_revenue = revenue_by_category.max()

print("Category with highest revenue:", top_category)
print("Revenue generated by that category:", top_category_revenue)

# Average revenue per sale
average_revenue = df["revenue"].mean()
print("Average revenue per sale:", average_revenue)

# Total revenue per quarter
df["quarter"] = df["date"].dt.to_period("Q").astype(str)
revenue_by_quarter = df.groupby("quarter")["revenue"].sum()

# Create a bar chart for quarterly revenue
plt.figure(figsize=(8, 5))
plt.bar(revenue_by_quarter.index, revenue_by_quarter.values)
plt.title("Total Revenue by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.show()
```

**üåü Exercise 5 : Data Visualisation Using MatPlotLib**

Instructions:
* Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
* Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
* Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.

```python
# Line plot: y = x^2 from -10 to 10
x = list(range(-10, 11))
y = [i**2 for i in x]

plt.figure()
plt.plot(x, y)
plt.title("Line Plot of y = x¬≤")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# Bar chart: Weekly product sales

products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.figure()
plt.bar(products, sales)
plt.title("Weekly Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()

# Pie chart: Favorite fruits

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
colors = ['red', 'yellow', 'pink', 'brown']

plt.figure()
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%')
plt.title("Favorite Fruits")
plt.legend()
plt.show()
```
