```python
# Install the libraries
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from scipy.ndimage import rotate
import numpy as np
```

```python
# To load an image and display it
image_path = 'Flower Color Images Dataset/flowers/flowers/19_010.png'
original_image = Image.open(image_path)

plt.imshow(original_image)
plt.axis('off')
plt.show()
```

```python
# Rotate an image by 30 degrees
def rotate_image_30_degrees(image):
    return rotate(np.array(image), 30, reshape=False, mode='nearest')

rotated_image = rotate_image_30_degrees(original_image)

plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

```python
# Flip an image horizontally and then vertically using mirror and flip functions from ImageOps
horizontal_flip = ImageOps.mirror(original_image)
vertical_flip = ImageOps.flip(original_image)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Horizontal flip")
plt.imshow(horizontal_flip)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Vertical flip")
plt.imshow(vertical_flip)
plt.axis('off')

plt.show()
```

```python
# Zoom in on an image (scale by 1.2x) using .resize(...)
w, h = original_image.size
zoomed_image = original_image.resize((int(w * 1.2), int(h * 1.2)))

plt.imshow(zoomed_image)
plt.axis('off')
plt.show()
```
